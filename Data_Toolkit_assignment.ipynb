17) Explain the difference between apply() and map() in Pandas.
sol) In Pandas, both apply() and map() are used to transform data within a DataFrame or Series, but they serve different purposes and have different use cases.

map()
Series-specific: The map() method is specifically designed for a Series (a single column of a DataFrame).

Element-wise operation: It operates on each element of the Series independently. It's often used for mapping values from one set to another, like replacing categorical values with numerical ones.

Efficiency: It is highly optimized for this specific task and is generally faster than apply() for element-wise transformations on a Series.

Syntax: Series.map(arg) where arg can be a function, a dictionary, or a Series.

apply()
Series and DataFrame: The apply() method can be used on both a Series and a DataFrame.

Row/Column or Element-wise: When used on a DataFrame, it applies a function along a specific axis (rows or columns). This means it can perform operations that require the context of an entire row or column, such as calculating the sum of a row or finding the maximum value.

Flexibility: It is more flexible than map() and can handle more complex operations, including those that return multiple values. However, this flexibility can come with a performance cost.

Syntax: DataFrame.apply(func, axis=...) or Series.apply(func).

Analogy: Think of map() as a translation dictionary for a single column, where each word is looked up and replaced. Think of apply() as a function machine that takes an entire row or column as input and performs a calculation on it, like summing up all the numbers in a row.

21) Why is NumPy's array slicing faster than Python's list slicing?
sol) NumPy's array slicing is significantly faster than Python's list slicing primarily due to how they are implemented and how they handle memory.

NumPy Array Slicing
View, not Copy: NumPy slicing creates a view of the original array. This means that the new "slice" is just a reference to a specific portion of the original data in memory. No new data is copied, which is extremely fast and memory-efficient, especially for large arrays.

Contiguous Memory: As mentioned before, NumPy arrays are stored in a contiguous block of memory. This allows for very fast access to data because the processor can predict where the next element is. Slicing simply involves calculating the start and end pointers in that block.

Python List Slicing
Creates a Copy: When you slice a Python list, it creates a new list object and copies the elements from the original list into the new one. This copying process takes time and consumes additional memory.

Non-contiguous Memory: Python lists are not stored contiguously in memory; they are a collection of pointers to objects located elsewhere. Slicing requires iterating through these pointers and copying the referenced objects, which is a much slower process.

Summary: NumPy's slicing is faster because it avoids data copying, operating on views of memory, which is possible due to its contiguous memory allocation. In contrast, list slicing is slower because it always creates a new list by copying the elements.

4) How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

```python
# This script calculates a correlation matrix from a DataFrame
# and visualizes it as a heatmap using Seaborn.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create a sample DataFrame with numerical data.
# The `corr()` method works on numerical columns.
data = {
    'Feature_A': [10, 20, 30, 40, 50],
    'Feature_B': [1, 2, 3, 4, 5],
    'Feature_C': [5, 4, 3, 2, 1],
    'Feature_D': [100, 105, 110, 115, 120]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("-" * 30)

# 2. Calculate the correlation matrix.
# The .corr() method computes the pairwise correlation of columns.
correlation_matrix = df.corr()

print("Correlation Matrix:")
print(correlation_matrix)
print("-" * 30)

# 3. Create the heatmap for visualization.
# We pass the correlation matrix to sns.heatmap().
# 'annot=True' displays the correlation values on the heatmap.
# 'cmap' sets the color map (e.g., 'coolwarm' is a good choice for this).
plt.figure(figsize=(8, 6)) # Adjust figure size for better display
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')

# 4. Display the plot.
plt.show()
```

11) Perform matrix multiplication using NumPy

```python
# This script demonstrates how to perform matrix multiplication using the NumPy library.
# Matrix multiplication is a fundamental operation in linear algebra and is used extensively
# in data science, machine learning, and scientific computing.

import numpy as np

# 1. Define two matrices (NumPy arrays).
# Matrix A must have dimensions (m, n) and Matrix B must have dimensions (n, p)
# for the multiplication to be possible. The resulting matrix will have dimensions (m, p).
# Here, Matrix A is (2, 3) and Matrix B is (3, 2).

# Create Matrix A
matrix_a = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Create Matrix B
matrix_b = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)
print("-" * 30)

# 2. Perform matrix multiplication.
# NumPy offers two primary ways to perform matrix multiplication:
# a) The @ operator (recommended for clean syntax)
result_matrix = matrix_a @ matrix_b

# b) The np.dot() function
# result_matrix = np.dot(matrix_a, matrix_b)

# 3. Print the result of the multiplication.
print("Result of Matrix A @ Matrix B:")
print(result_matrix)

# You can verify the dimensions of the result.
# The result will be a (2, 2) matrix.
print("\nShape of the result matrix:", result_matrix.shape)
```

```python
# This script demonstrates how to use the pandas library to load a CSV file
# and display the first few rows of the resulting DataFrame.

import pandas as pd
import os # We'll use this to clean up the dummy file

# 1. Create a dummy CSV file for this example.
# In a real-world scenario, you would already have a CSV file.
# We'll create a DataFrame and save it to a file.
data = {
    'Product': ['Apple', 'Banana', 'Orange', 'Grape', 'Strawberry', 'Blueberry', 'Mango', 'Pineapple'],
    'Price': [1.20, 0.75, 0.90, 2.50, 3.10, 2.95, 1.50, 2.00],
    'Quantity_Sold': [150, 200, 180, 50, 90, 110, 130, 85]
}
df_dummy = pd.DataFrame(data)
dummy_file_name = 'sample_data.csv'
df_dummy.to_csv(dummy_file_name, index=False)
print(f"Created a dummy CSV file: '{dummy_file_name}'\n")

# 2. Load the CSV file into a pandas DataFrame.
# The pd.read_csv() function is the primary tool for this.
try:
    df = pd.read_csv(dummy_file_name)

    # 3. Display the first 5 rows of the DataFrame.
    # The .head() method is used for this purpose. It's great for
    # getting a quick overview of your data's structure.
    print("DataFrame loaded from CSV. First 5 rows:")
    print(df.head())

except FileNotFoundError:
    print(f"Error: The file '{dummy_file_name}' was not found.")

finally:
    # Clean up the dummy file.
    if os.path.exists(dummy_file_name):
        os.remove(dummy_file_name)
        print(f"\nCleaned up dummy file: '{dummy_file_name}'")
```

```python
# This script demonstrates how to create an interactive 3D scatter plot using Plotly.
# 3D scatter plots are useful for visualizing relationships between three variables.

import pandas as pd
import numpy as np
import plotly.express as px

# 1. Create a sample dataset.
# We'll create a DataFrame with three numerical columns (x, y, z) and one
# categorical column ('group') to color the points.
np.random.seed(42)  # For reproducibility
data = {
    'x': np.random.randn(100),
    'y': np.random.randn(100),
    'z': np.random.randn(100),
    'group': np.random.choice(['A', 'B', 'C'], 100)
}
df = pd.DataFrame(data)

# 2. Create the 3D scatter plot using plotly.express.
# This function provides a simple interface to create the plot.
# - The `x`, `y`, and `z` arguments specify the columns for each axis.
# - The `color` argument is used to differentiate points based on a categorical variable.
fig = px.scatter_3d(df, x='x', y='y', z='z',
                    color='group',
                    title='Interactive 3D Scatter Plot')

# 3. Show the plot.
# This will open a new window or display the plot in a notebook
# with interactive features like zooming, panning, and rotating.
fig.show()
```
